# Basics of Data Visualization

Data visualization is the process of **representing data visually** to understand patterns, trends, and relationships.  

Key principles:
- Choose the right type of plot for your data (categorical vs numeric).  
- Label axes and provide a title.  
- Use color and shapes to highlight differences.  
- Keep it simple and clear for interpretation.

Common plots in R:
- Bar plot: For categorical data
- Histogram: For numeric data distribution
- Boxplot: For spread and outliers
- Scatter plot: For relationships between two numeric variables


```r
# Example dataset
ages <- c(22, 25, 30, 28, 35, 40, 23, 27, 31, 29)
heights <- c(160, 165, 175, 170, 180, 185, 162, 168, 172, 169)
```

```r
# Bar plot (categorical example)
categories <- c("A", "B", "C", "A", "B", "C", "A", "B", "C", "A")
category_counts <- table(categories)
```

```r
par(bg = "white", col.axis = "black", col.lab = "black", col.main = "black", pch = 19)
barplot(category_counts, col = "orange", main = "Category Counts",
        xlab = "Categories", ylab = "Frequency")
```

```r
# Histogram (numeric example)
par(bg = "white", col.axis = "black", col.lab = "black", col.main = "black", pch = 19)
hist(ages, col = "skyblue", main = "Age Distribution",
     xlab = "Age", ylab = "Frequency", border = "white")
```

```r
# Boxplot (to visualize spread and outliers)
par(bg = "white", col.axis = "black", col.lab = "black", col.main = "black", pch = 19)
boxplot(heights, col = "lightgreen", main = "Height Spread",
        ylab = "Height (cm)")
```

```r
# Scatter plot (relationship between age and height)
par(bg = "white", col.axis = "black", col.lab = "black", col.main = "black", pch = 19)
plot(ages, heights, col = "purple", pch = 19,
     main = "Age vs Height", xlab = "Age", ylab = "Height (cm)")
abline(lm(heights ~ ages), col = "red", lwd = 2) # regression line
```
